```python
df_user=pd.read_csv('BX-Users.csv',encoding='latin-1', low_memory=False)
df_ratings=pd.read_csv('BX-Book-Ratings.csv',encoding='latin-1', low_memory=False,nrows=10000)
df_book=pd.read_csv('BX-Books.csv',encoding='latin-1', low_memory=False)
```

```python
df_rating_book= df_ratings.merge(df_book, on='isbn')
df_rating_book
```

```python
n_users=df_rating_book.user_id.nunique()
n_books=df_rating_book.isbn.nunique()

print("Number of user:", n_users)
print("Number of books:", n_books)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_rating_book['User_id_order']=le.fit_transform(df_rating_book['user_id'])
df_rating_book['isbn_order']=le.fit_transform(df_rating_book['isbn'])
df_rating_book.head()
```

```python
isbn_list = df_rating_book.isbn.unique()
print(" Length of isbn List:", len(isbn_list))
def get_isbn_numeric_id(isbn):
    #print ("  isbn is:" , isbn)
    itemindex = np.where(isbn_list==isbn)
    return itemindex[0][0]
```

```python
userid_list = df_rating_book.user_id.unique()
print(" Length of user_id List:", len(userid_list))
def get_user_id_numeric_id(user_id):
    #print ("  isbn is:" , isbn)
    itemindex = np.where(userid_list==user_id)
    return itemindex[0][0]
```

```python
df_rating_book['user_id_order'] = df_rating_book['user_id'].apply(get_user_id_numeric_id)
```

```python
df_rating_book['isbn_id'] = df_rating_book['isbn'].apply(get_isbn_numeric_id)
df_rating_book.head()
```

```python
new_col_order = ['user_id_order', 'isbn_id', 'rating', 'book_title', 'book_author','year_of_publication','publisher','isbn','user_id']
df_rating_book = df_rating_book.reindex(columns= new_col_order)
df_rating_book.head()
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(df_rating_book, test_size=0.30)
```

```python
train_data_matrix=np.zeros((n_users,n_books))
for line in train_data.itertuples():
    train_data_matrix[line[1]-1,line[2]-1]=line[3]
test_data_matrix=np.zeros((n_users,n_books))
for line in test_data.itertuples():
    test_data_matrix[line[1]-1,line[2]-1]=line[3]
```

```python
from sklearn.metrics import pairwise_distances
user_similar=pairwise_distances(train_data_matrix,metric='cosine')
item_similar=pairwise_distances(train_data_matrix.T,metric='cosine')
```

```python
user_similar
```

```python
def predict(ratings, similarity, type='user'):
    if type=='user':
        mean_user_rating=ratings.mean(axis=1)
        rating_diff= (ratings-mean_user_rating[:,np.newaxis])
        pred=mean_user_rating[:,np.newaxis]+ similarity.dot(rating_diff)/np.array([np.abs(similarity).sum(axis=1)]).T
    elif type=='item':
        pred= ratings.dot(similarity)/ np.array([np.abs(similarity).sum(axis=1)])
    return pred
```

```python
user_prediction = predict(train_data_matrix, user_similar, type='user')
item_prediction = predict(train_data_matrix, item_similar, type='item')
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(prediction, ground_truth):
    prediction = prediction[ground_truth.nonzero()].flatten() 
    ground_truth = ground_truth[ground_truth.nonzero()].flatten()
    return sqrt(mean_squared_error(prediction, ground_truth))
```

```python
print('User-based CF RMSE: ' + str(rmse(user_prediction, test_data_matrix)))
print('Item-based CF RMSE: ' + str(rmse(item_prediction, test_data_matrix)))
```
